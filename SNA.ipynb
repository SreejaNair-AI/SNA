```python
#Question A:
# Question 1: Load the data as directed graph. [2 marks]
import pandas as pd

data = pd.read_csv('/sreejarawfile.csv') # to load the data

# Initialize directed graph
G = nx.from_pandas_edgelist(data, 'src_node', 'dst_node', create_using=nx.DiGraph)
G = nx.Graph()
print(data.head())
print(data.columns)


#Question 2: Draw the graph. [3 marks]

import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add edges from the DataFrame using the correct columns
for index, row in data.iterrows():
    src = row['src_node']  # Source node
    dst = row['dst_node']  # Destination node
    G.add_edge(src, dst)

# Check if any nodes or edges were added
print("Nodes in graph:", G.nodes())
print("Edges in graph:", G.edges())

# Now plot the graph
if G.number_of_nodes() > 0 and G.number_of_edges() > 0:
    pos = nx.spring_layout(G)  # Set layout for better visualization
    plt.figure(figsize=(10, 6))
    nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_color='black', font_weight='bold', edge_color='gray')
    plt.title("Directed Graph of Nodes")
    plt.show()
else:
    print("The graph has no nodes or edges.")


#Question 3: Draw the degree distribution for in-degree and out-degree separately. [5 marks]

in_degree = dict(G.in_degree())
out_degree = dict(G.out_degree())

plt.figure(figsize=(10,5))

# Plotting in-degree distribution
plt.subplot(1, 2, 1)
plt.hist(in_degree.values(), bins=10)
plt.title('In-degree Distribution')
plt.xlabel('In-degree')
plt.ylabel('Frequency')

# Plotting out-degree distribution
plt.subplot(1, 2, 2)
plt.hist(out_degree.values(), bins=10)
plt.title('Out-degree Distribution')
plt.xlabel('Out-degree')
plt.ylabel('Frequency')

plt.show()


#Question B:

#Question "1. How many strongly connected components are there? [1 marks]\n",
scc = list(nx.strongly_connected_components(G))
print(f"Number of strongly connected components: {len(scc)}")

#Question "2. Remove all nodes and edges that are not part of any strongly connected components. [4 marks]\n",
largest_scc = max(scc, key=len)
H = G.subgraph(largest_scc).copy()

# Question"3. Draw the graph after node+edge removal. [3 marks]\n",
plt.figure(figsize=(10,10))
nx.draw(H, with_labels=True, node_size=500, node_color="lightgreen", font_size=10, font_color="black", edge_color="gray")
plt.show()


#Question 4. Is this graph connected? If not, how many components are there? [2 marks]"
is_connected = nx.is_strongly_connected(H)
print(f"Is the graph connected? {is_connected}")


#Question C:
#"1. Load the graph with only friendship relationship as undirected graph. **Note that friendship relations are those which has resiprocal links. You should not load any edge that is not having any resiprocal edges in the given data set.** [2 marks]\n",
G_undirected = nx.Graph([(u, v) for u, v in G.edges() if G.has_edge(v, u)])

#"2. Draw the new graph. [2 marks]\n",
plt.figure(figsize=(10,10))
nx.draw(G_undirected, with_labels=True, node_size=500, node_color="lightcoral", font_size=10, font_color="black", edge_color="gray")
plt.show()

#"3. Draw the degree distribution. [1 marks]"
degree = dict(G_undirected.degree())
plt.figure(figsize=(10,5))
plt.hist(degree.values(), bins=10)
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.show()

#Question D:
#"1. With the friends graph generated in the previous step, find all the pair of nodes which has distance 2 hop but no direct connections. [2 marks]\n",
# Finding all pairs of nodes with distance 2 and no direct connection
pairs_distance_2 = []

# Iterate through all nodes
for node in G.nodes():
    # Get neighbors of the current node
    neighbors = set(G.neighbors(node))

    # Find all nodes that are two hops away
    for neighbor in neighbors:
        for second_neighbor in G.neighbors(neighbor):
            if second_neighbor != node and second_neighbor not in neighbors:  # not directly connected
                pairs_distance_2.append((node, second_neighbor))

# Remove duplicates
pairs_distance_2 = list(set(pairs_distance_2))
print("Pairs with distance 2 and no direct connection:", pairs_distance_2)

#"2. For all these pair of nodes calculate Jacard Coefficient, Preferential attachment score, Adamic-Adar Index, and resource allocation index. [2 marks]\n",
G_undirected = nx.to_undirected(G)
# Calculate scores for each pair
scores = []
for u, v in pairs_distance_2:
    jaccard_coefficient = list(nx.jaccard_coefficient(G_undirected, [(u, v)]))[0][2]  # get the score from the generator
    preferential_attachment = list(nx.preferential_attachment(G_undirected, [(u, v)]))[0][2]
    adamic_adar_index = list(nx.adamic_adar_index(G_undirected, [(u, v)]))[0][2]
    resource_allocation_index = list(nx.resource_allocation_index(G_undirected, [(u, v)]))[0][2]

    scores.append((u, v, jaccard_coefficient, preferential_attachment, adamic_adar_index, resource_allocation_index))

#"3. Rank the pairs based on the index scores calculated. [2 marks]\n",
for score in scores:
    print(f"Pair: {score[0]} - {score[1]} | Jaccard: {score[2]} | Preferential Attachment: {score[3]} | "
          f"Adamic-Adar: {score[4]} | Resource Allocation: {score[5]}")

#"4. 5 separate ranks will be there. Top (highest score) are the predicted links. Analyze the differences in the top 10 predicted links by different methods. Discuss your findings. [4 marks]"

#Question E:
#"1. Detect communities using any suitable methods available in networkx. [2 marks]\n",
#"2. Let us consider a metric for verification if for any two nodes u and v inside the detected community is from the same space (one node may be in multiple space, any one match will be okay) is true positive (TP). If they are from distinct space, it is false positive (FP). If two nodes from different community is from same space, it is false negative (FN) and if two nodes from different communities are from different space then it true negative (TN). The accuracy measure is $$ \\frac{TP + TN}{TP+FP+TN+FN}$$ Find the accuracy for the method you used. [3 marks]"
import networkx as nx
communities = nx.algorithms.community.girvan_newman(G)
first_communities = next(communities)
for i, community in enumerate(first_communities):
    print(f"Community {i + 1}: {community}")

#let us define the node space
node_space = {
    1: "Space_A",
    2: "Space_A",
    3: "Space_B",
    4: "Space_C",
}
# Initialize counts
TP = FP = FN = TN = 0
for community in first_communities:
    for u in community:
        for v in community:
            if u != v:
                if u in node_space and v in node_space:
                    if node_space[u] == node_space[v]:
                        TP += 1  # True Positive
                    else:
                        FP += 1  # False Positive

# For nodes in different communities
all_nodes = set(G.nodes())
for community in first_communities:
    other_communities = [c for c in first_communities if c != community]
    for u in community:
        for other_community in other_communities:
            for v in other_community:
                # Check if both nodes exist in the node_space dictionary
                if u in node_space and v in node_space:
                    if node_space[u] == node_space[v]:
                        FN += 1  # False Negative
                    else:
                        TN += 1  # True Negative

# Calculate accuracy
accuracy = (TP + TN) / (TP + FP + TN + FN) if (TP + FP + TN + FN) > 0 else 0

print(f"Accuracy: {accuracy:.2f}")
```
